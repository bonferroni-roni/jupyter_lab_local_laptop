# Lecture 9 - XGBoost with SHAP in R on COVID data

## Begin with loading packages we need for this example

```r
library(caret)
library(splitTools)
library(ROSE)
library(rpart)
library(caret)
require(xgboost)
```

## Set up the data with complete rows

```r
df <- sdgd::oncovid
df <- df[complete.cases(df),]
df$case_status <- as.factor(df$case_status)
head(df)
```

## Split the data into 60/40 train/test proportions

```r
# split to 60% training and 40% testing
inds <- splitTools::partition(df$case_status, p = c(train = 0.6, test = 0.4))
```

## Balance the training data

```r
# balance the training data
df_train <- 
  ROSE::ovun.sample(case_status ~., data=df[inds$train,], p=0.5, seed=11, method="under")$data
```

## For XGBoost, we need to convert the data frame into a matrix

```r
dt <- data.matrix(df_train[,-1])
lbl <- as.numeric(df_train$case_status)-1 # as.numeric convert to1 and 2 instead of 1 and 0!!!
```

## Run the XGBoost 

### in verbose mode (to see it work) and with "binary:logistic" to produce the classification scores

```r
bst <- 
  xgboost::xgboost(data= dt, label= lbl , max.depth= 2, eta= 1, nthread= 2, nrounds= 5, 
          objective= "binary:logistic", verbose = 1, predcontrib = TRUE)
```

## Obtain the predictions on the test data

### Note: use xgb.DMatrix to convert the data frame into a matrix first

```r
inds_sample <- sample(inds$test, size= 0.1 *length(inds$test),replace=T)

# obtain predictions from the bag of tree models and average the results
ts <- xgboost::xgb.DMatrix(data = data.matrix(df[inds_sample,-1]), label= as.numeric(df[inds_sample,1]))
p1 <- predict(bst, ts)
# p1 <- predict(bst, ts, predcontrib = TRUE, approxcontrib = F)
head(p1)
```

## Determine classifications using 0.5 decision threshold then examine the performance

```r
pred <- ifelse(p1>0.5, 1, 0)
caret::confusionMatrix(as.factor(pred), as.factor(df[inds_sample,"case_status"]))
```

## Use 0.5 for the classification decision threshold

### then examine the performance …..

```r
pred <- ifelse(p1>0.7, 1, 0)
    caret::confusionMatrix(as.factor(pred), as.factor(df[inds_sample,"case_status"]))

# calculate the AUC
auc <- sdgm::auc(pred, df[inds_sample,"case_status"])

cat("The AUC is: ", auc)
```

## Obtain SHAP Values for each feature

```r
xgboost::xgb.plot.shap(data = data.matrix(df[inds$test,-1]), model = bst, top_n = 5)

xgboost::xgb.plot.shap.summary(data = data.matrix(df[inds$test,-1]), model = bst)

shap_values <- 
   xgboost::xgb.plot.shap(data = data.matrix(df[inds$test,-1]), model = bst, features=colnames(dt), plot = FALSE)

head(shap_values$shap_contrib)
```

## SHAPLEY Values for each data point

### For this section, we need to reduce the size of the test data!!!!!!!

```r
inds_sample <- sample(inds$test, size= 0.1 *length(inds$test),replace=F)
```

## Prepare the data for explanation

```r
library(shapr)

# use the training matrix dt and the boosted model bst from above
explainer <- shapr::shapr(dt, bst)
p <- mean(predict(bst, dt))
```

## Computing the actual Shapley values with kernelSHAP accounting for feature dependence using the empirical (conditional) distribution approach


```r
explanation <- shapr::explain(
  data.matrix(df[inds_sample,-1]),
  approach = "empirical",
  explainer = explainer,
  prediction_zero = p
)
```

## Print the Shapley values for the test data.

```r
print(explanation$dt)
```

## Plot the resulting explanations for the data point on row 4 of test data

```r
plot(explanation, plot_phi0 = FALSE, index_x_test = c(4))
```

## Plot the resulting explanations for data points on rows 34 and 666 of the test data

```r
plot(explanation, plot_phi0 = FALSE, index_x_test = c(34, 26, 46, 29))
```
